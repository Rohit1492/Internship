# 1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’. 

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

def AllHeader(link):
    page= requests.get('https://en.wikipedia.org/wiki/Main_Page')
    soup=BeautifulSoup(page.content)

    header=[] #creating empty list for store data
    for i in soup.find_all('span', class_="mw-headline"):
        header.append(i.text)
    return header

AllHeader('https://en.wikipedia.org/wiki/Main_Page')
```

# 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

```python
# Send request to the webpage server and get the source code of the web page
imd= requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
imd
```

```python
from urllib.parse import urlparse
import requests
from bs4 import BeautifulSoup
import pandas as pd

def imdbTop100(link):
    
    movie100Title=[]
    movie100Rating=[]
    movie100Year=[]
    next50MovieLinkArray=[]
    parseLink = urlparse(link)
    imdb= requests.get(link)
    soup=BeautifulSoup(imdb.content)
    for i in soup.select('h3.lister-item-header>a'):
        movie100Title.append(i.text)
    for i in soup.find_all('div',class_="inline-block ratings-imdb-rating"):
        movie100Rating.append(i.text.strip())
    for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
        movie100Year.append(i.text.strip('(').strip(')'))
        
    next50MovieLinkArray=soup.select("div.desc>a")
    next50MovieLink = parseLink.scheme +'://'+ parseLink.netloc + next50MovieLinkArray[0]['href']
    
    imdb2=requests.get(next50MovieLink)
    soup2 = BeautifulSoup(imdb2.content)
    for i in soup2.select('h3.lister-item-header>a'):
        movie100Title.append(i.text)
    for i in soup2.find_all('div',class_="inline-block ratings-imdb-rating"):
        movie100Rating.append(i.text.strip())
    for i in soup2.find_all('span',class_="lister-item-year text-muted unbold"):
        movie100Year.append(i.text.strip('(').strip(')'))
    
    df= pd.DataFrame({'Title':movie100Title, 'Rating':movie100Rating,'Year':movie100Year})
    
    return df
    
    
```

```python
imdbTop100('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
```

# 3. Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def top100IndMovie(link):
    movieTitle=[]
    movieRating=[]
    movieYear=[]
    
    imdb=requests.get(link)
    soup=BeautifulSoup(imdb.content)
    for i in soup.select('td.titleColumn>a'):
        movieTitle.append(i.text)
    for i in soup.find_all('td',class_="ratingColumn imdbRating"):
        movieRating.append(i.text.strip())
    for i in soup.find_all('span', class_="secondaryInfo"):
        movieYear.append(i.text)
        
    df= pd.DataFrame({'Title':movieTitle[0:100],'Rating':movieRating[0:100],'Yaer':movieYear[0:100]})
        
    return df
```

```python
top100IndMovie('https://www.imdb.com/india/top-rated-indian-movies/?sort=ir,desc&mode=simple&page=1')
```

# 4. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have toscrape: 
i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating. 
ii) Top 10 ODI Batsmen in men along with the records of their team and rating. 
iii) Top 10 ODI bowlers along with the records of their team and rating.


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def top10Menteam(link):
    teams=[]
    points=[]
    rating=[]
    
    team=requests.get(link)
    soup=BeautifulSoup(team.content)
    
    for i in soup.find_all('span',class_='u-hide-phablet'):
        teams.append(i.text)
        
    matches1=[]    
    for i in soup.find_all('td',class_='rankings-block__banner--matches'):
        matches1.append(i.text)
    matches2=[]
    for i in soup.find_all('td',class_='table-body__cell u-center-text'):
        matches2.append(i.text)
    matches=matches1+matches2[0:18:2]
        
    for i in soup.find_all('td',class_='rankings-block__banner--points'):
        points.append(i.text)
    for i in soup.find_all('td',class_='table-body__cell u-center-text'):
        points.append(i.text)
        
    for i in soup.find_all('td',class_='rankings-block__banner--rating u-text-right'):
        rating.append(i.text.strip()) 
    for i in soup.find_all('td',class_='table-body__cell u-text-right rating'):
        rating.append(i.text.strip())
    
    df= pd.DataFrame({'Team':teams[0:10],'Matches':matches,'Points':points[0:20:2],'Rating':rating[0:10]})
    return df
    
```

```python
top10Menteam('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

# 8.Write a python program to scrape mentioned details from ‘https://www.dineout.co.in/delhi-restaurants/buffetspecia

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# i) Restaurant name
# ii) Cuisine
# iii) Location
# iv) Ratings
# v) Image url

def restaurent(link):
    Restaurantname=[]
    Cuisine=[]
    Location=[]
    Ratings=[]
    ImageURL=[]
    
    rest=requests.get(link)
    soup=BeautifulSoup(rest.content)
    
    for i in soup.find_all('a', class_="restnt-name ellipsis"):
        Restaurantname.append(i.text)
    for i in soup.find_all('div', class_="restnt-loc ellipsis"):
        Location.append(i.text)
    for i in soup.find_all('img', class_="no-img"):
        ImageURL.append(i['data-src'])
    for i in soup.find_all('span', class_="double-line-ellipsis"):
        Cuisine.append(i.text.split("|")[1])
    for i in soup.find_all('div', class_="restnt-rating"):
        Ratings.append(i.text)
    
    df= pd.DataFrame({'Title':Restaurantname,'Cuisine':Cuisine,'Location':Location,'Rating':Ratings[0:13],'ImageURL':ImageURL})
    return df
    
    
    
    
```

```python
restaurent('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

# 9.Write a python program to scrape weather details for last 24 hours from
‘https://en.tutiempo.net/delhi.html?data=last-24- hours’ :

```python
# i) Hour
# ii) Temperature
# iii) Wind
# iv) Weather condition
# v) Humidity
# vi) Pressure

import requests
from bs4 import BeautifulSoup
import pandas as pd

def Weather(link):
    Temperature=[]
    Wind=[]
    Weathercondition=[]
    Humidity=[]
    Pressure=[]
    
    weather=requests.get(link)
    soup=BeautifulSoup(weather.content)
    
    hour=[]
    for i in soup.select('div.last24.thh>table>tbody>tr>td'):
        hour.append(i.text)
    Hour=hour[:144:6]
    
    for i in soup.find_all('span',class_='thhip ico i0530 u303n'):
        Weathercondition.append(i.text)    
    for i in soup.find_all('td',class_='t Temp'):
        Temperature.append(i.text)    
    for i in soup.find_all('td',class_='wind'):
        Wind.append(i.text)
    for i in soup.find_all('td',class_='hr'):
        Humidity.append(i.text)
    for i in soup.find_all('td',class_='prob'):
        Pressure.append(i.text)
    
    df=pd.DataFrame({'hour':Hour,'temperature':Temperature[0:24],'wind':Wind[0:24],'weather':Weathercondition[0:24],'humidity':Humidity[0:24],'pressure':Pressure[0:24]})

    return df

```

```python
Weather('https://en.tutiempo.net/delhi.html?data=last-24-hours')
```

# 10.Write a python program to scrape monument name, monument description, image url about top 10 monuments
from 'https://www.puredestinations.co.uk/top-10-famous-monuments-to-visit-in-india/'

```python

import requests
from bs4 import BeautifulSoup
import pandas as pd

def monument(link):
    name=[]
    desc=[]
    image=[]
    filteredDesc=[]
    
    monu=requests.get(link)
    soup=BeautifulSoup(monu.content)
    
    for i in soup.find_all('strong'):
        name.append(i.text)
    for i in soup.select('div.blog--single__content>p'):
        desc.append(i.text)
    for i in soup.select('div.blog--single__content>p>img'):
        image.append(i['data-srcset'])
        
    for ele in desc:
        if ele!='':
            filteredDesc.append(ele)
    dd = filteredDesc[1:]
        
    df= pd.DataFrame({'name':name[0:10],'description':dd[1:20:2],'image':image})
    return df
```

```python
monument('https://www.puredestinations.co.uk/top-10-famous-monuments-to-visit-in-india/')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def House(link):
    title=[]
    loc=[]
    area=[]
    emi=[]
    price=[]
    
    page=requests.get(link)
    soup=BeautifulSoup(page.content)
    
    for i in soup.select('h2.heading-6.font-semi-bold.nb__1AShY>span'):
        title.append(i.text)
        
    for i in soup.find_all('div', class_='nb__2CMjv'):
        loc.append(i.text)
    
    for i in soup.find_all('div', class_='nb__3oNyC'):
        area.append(i.text)
    
    for i in soup.find_all('div', id='roomType'):
        emi.append(i.text)
    
    for i in soup.find_all('div',class_='font-semi-bold heading-6'):
        price.append(i.span)    
    Price=[]
    for x in price:
        if x!=None:
            Price.append(x.text)
            
    df= pd.DataFrame({'House':title,'Location':loc,'Area':area,'Emi':emi,'Price':Price[1::2]})
    
    return df
```

# 5. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have toscrape:
i) Top 10 ODI teams in women’s cricket along with the records for matches, points andrating.
ii) Top 10 women’s ODI players along with the records of their team and rating.
iii) Top 10 women’s ODI all-rounder along with the records of their team andrating.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def top10Womenteam(link):
    teams=[]
    rating=[]
    
    team=requests.get(link)
    soup=BeautifulSoup(team.content)
    
    for i in soup.find_all('span',class_='u-hide-phablet'):
        teams.append(i.text)
    
    m1=[]    
    for i in soup.find_all('td',class_='rankings-block__banner--matches'):
        m1.append(i.text)
    m2=[]    
    for i in soup.find_all('td',class_='table-body__cell u-center-text'):
        m2.append(i.text)
    matches=m1+m2[0:18:2]
    
    p1=[]    
    for i in soup.find_all('td',class_='rankings-block__banner--points'):
        p1.append(i.text)
    p2=[]    
    for i in soup.find_all('td',class_='table-body__cell u-center-text'):
        p2.append(i.text)
    points=p1+p2[1:20:2]
    
    r1=[]
    for i in soup.find_all('td',class_='rankings-block__banner--rating u-text-right'):
        r1.append(i.text.strip()) 
    r2=[]
    for i in soup.find_all('td',class_='table-body__cell u-text-right rating'):
        r2.append(i.text.strip())
    rating=r1+r2

    
    df= pd.DataFrame({'Team':teams[0:10],'Matches':matches,'Points':points,'Rating':rating})
        
    return df
```

```python
top10Womenteam('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
```

# ii) Top 10 women’s ODI players along with the records of their team and rating. 

```python
def top10WoMenPlayer(link):
    rating=[]
    
    women=requests.get(link)
    soup=BeautifulSoup(women.content)
    
    bm1=[]
    for i in soup.find('div',class_='rankings-block__banner--name-large'):
        bm1.append(i)
    bm2=[]
    for i in soup.select('td.table-body__cell.rankings-table__name.name>a'):
        bm2.append(i.text)
    BatsWomen=bm1+bm2[0:9]
    
    
    for i in soup.find('div', class_='rankings-block__banner--rating'):
        rating.append(i)
    for i in soup.find_all('td',class_='table-body__cell rating'):
        rating.append(i.text)
        
    team=[]
    for i in soup.find_all('div',class_='rankings-block__banner--nationality'):
        team.append(i.text.strip())
    
    for i in soup.find_all('span',class_='table-body__logo-text'):
        team.append(i.text)
    
            
    df= pd.DataFrame({'Player':BatsWomen,'Team':team[0:10],'Rating':rating[0:10]})
    
    return df
```

```python
top10WoMenPlayer('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')
```
